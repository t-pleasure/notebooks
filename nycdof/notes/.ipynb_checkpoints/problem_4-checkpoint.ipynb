### General Approach

#### Main Idea:
* Since price has been specified by the tiering, this analysis will only look at growth in terms of number of sales for a particular price tier. To do so, compute monthly year-over-year growth (i.e., how much the total sales from year[x] differs from year[x-1] as a percentage) for each month/year.

#### Steps:
1. Group rows by (month, year, tier)
2. Compute total number of sales for each (month, year, tier).
3. Compute the growth for each (month, year, tier) relative to its previous year as: **(prev-cur)/prev**
4. Plot tiers on the same graph with the x-axis being time and y-axis being growth.
5. Group rows by neighborhoods (specifically brooklyn heights and bedford stuy), then repeat 2-4

## Data Analysis

```python
# import necessary libraries
%matplotlib inline
import csv
import pandas as pd
import datetime as dt
import seaborn as sns
from pylab import *
from collections import defaultdict

sns.set(context="talk", style="white")
```

### Reading and cleaning data

#### We will need to:
1. **remove** rows with $0 sale price (these are "transfer of ownership" events)

```python
# Tier classification function
def tier_3m(price):
  """
  Determines the tier of a price. Returns the appropriate string based
  on whether or not the price is geq or lesser than $3M
  Input:
   * price (number) - input price
  Output:
   * label (str) indicating price tier.
  """
  return "lessThan3m" if price < 3000000 else "moreThan3m"


# read csv data to data frame
df = pd.read_csv("../clean_sales_data.csv")

# add additional fields and rename fields to make life easier
df["neighborhood"] = df["NEIGHBORHOOD"]
df["sale_price"] = df["SALE PRICE"]
df["sale_date"] = df["SALE DATE"].map(lambda s: dt.datetime.strptime(s, "%m/%d/%Y"))
df["sale_day"] = df.sale_date.map(lambda d: d.day)
df["sale_month"] = df.sale_date.map(lambda d: d.month)
df["sale_year"] = df.sale_date.map(lambda d: d.year)
df["price_tier"] = df["sale_price"].map(tier_3m)

# remove columns that are not needed
df_slice = df[["neighborhood","sale_price","price_tier","sale_date","sale_day","sale_month","sale_year"]]

## PERFORM FILTERING ##
# 1. remove transfer of ownership rows
df_slice = df_slice[df_slice.sale_price > 0]
```

### Aggregating Data
This section will compute monthly YoY growth based on tiers

```python
def monthly_yoy_growth(fr, min_year=2003, max_year=2014):
  """
  Given a DataFrame, computes monthly YoY growth for each (tier, month, year) relative
  to its prior year. This function will return a generator that will produce:
  {"tier": str, "date": datetime, "growth": float}
  
  Input:
    * fr (pandas.DataFrame) -- frame to aggregate over
    * min_year - minimum year to start evaluation
    * max_year - maximum year to perform evaluation
  Output:
    * generator <dict> - produces dictionaries indicating sales count growth for a 
                         particular date/year relative to its sales count form the previous year.
  """
  # group frame by (tier, month, year)
  grp = fr.groupby(["price_tier","sale_month","sale_year"])["sale_price"].count()
  # convert grp from [((tier,month,year),cnt)]* to a map: (tier, month) -> [(year, cnt)]*
  agg = defaultdict(list)
  for ((tier, mn, yr), cnt) in grp.iteritems():
    agg[(tier,mn)].append((yr, cnt))    
  # for each (tier, month) compute the YoY differences
  for ((tier,mn),tups) in agg.items():
    d = dict(tups)
    for yr in range(min_year+1,max_year+1):
      cur = d[yr] if yr in d else 0 # np.NAN
      prev = d[yr-1] if (yr-1) in d else np.NAN
      growth = (cur - prev)/(prev + 0.0)
      yield {"tier": tier, "date": dt.datetime(yr,mn,1), "growth": growth}
             
# obtain YoY growth data frame
yoyfr = pd.DataFrame(list(monthly_yoy_growth(df_slice))).sort("date")
# remove 12/2014 since we don't have data for it
yoyfr = yoyfr[yoyfr["date"] < dt.datetime(2014,12,1)]
```

### Visualization

```python
def pretty_plot(df, graph_title):
  """
  function to help plot multi-tiers
  
  Input:
    * df (pandas.DataFrame) -- frame to plot. Note: one column must contain date time
    * graph_title (str) -- title to display on graph.
  Output:
    * None
  """
  def plot(x, y, data=None, label=None, **kwargs):
    data.plot(x, y, label=label, rot=0, lw=3, grid=False, **kwargs)

  g = sns.FacetGrid(df, hue="tier", size=12,
                  legend_out=False)
  g.map_dataframe(plot, "date", "growth")
  ax = g.axes[0, 0]
  ax.set_title(graph_title, size=14)
  g.fig.tight_layout()
  ax.legend(loc=(.8, .05))

pretty_plot(yoyfr, "monthly YoY Growth!")
```

### What gives with all the NaN and the funky plots?
When computing **growth**, the following equation is used:
* growth = (cur_value - prev_value)/prev_value

Problems exist when *prev_value* is 0! In particular, dividing any number by 0 will result in a division by zero error. 

**What this means** is that we have (year,month,tiers) with zero sales record for particular neighborhoods!

We currently handle this by specifying growth to be NaN if prev_value is 0. As a result, our plots will not have points defined for instances where an NaN value is present. 

Furthermore, one may try to solve this problem by adding a small pertubation to the denominator of the equation. However, this may not have the desired result. For example, (100-0)/(0+0.0001) does not make any more sense than NaN.

#### Why are there zero sales for these neighborhoods for a particular (year/month)?
That's just how the data is at the moment unfortunately.
The following will show how sparse our data is for the greater than $3M tier for Bedford Stuyvesant.

## Appendix: Additional tiering graphs 
Our main anaylysis was between properties less than \$3M and properties greater than \$3M. In this section, we will briefly look at plots for different price tier structures.


```python
##
# Additional tier classification functions
##
def tier_mils(price):
  """
  Determines the tier of a price based on the following ranges"
  $1-$1M
  $1M-$3M
  $3M+
  Input:
   * price (number) - input price
  Output:
   * label (str) indicating price tier.
  """
  if price < 1000000:
    return "1-1M"
  elif price >= 1000000 and price < 3000000:
    return "1M-3M"
  elif price >= 3000000:
    return "3M+"
  raise Exception(price)
    
    
def tier_granular(price):
  """
  Determines the tier of a price based on the following ranges"
  $1-$200k
  $200k-$600k
  $600k-$1M
  $1M-$2M
  $2M-$3M
  $3M+
  Input:
   * price (number) - input price
  Output:
   * label (str) indicating price tier.
  """
  if price < 200000:
    return "1-200k"
  elif price >= 200000 and price < 600000:
    return "200k-600k"
  elif price >= 600000 and price < 1000000:
    return "600K-1M"
  elif price >= 1000000 and price < 2000000:
    return "1M-2M"
  elif price >= 2000000 and price < 3000000:
    return "2M-3M"
  elif price >= 3000000:
    return "3M+"
  raise Exception(price)
    
    
```

```python
df_slice["price_tier"] = df_slice.sale_price.map(tier_mils)
fr = pd.DataFrame(list(monthly_yoy_growth(df_slice)))
pretty_plot(fr, "yoy monthly growth using tier_mils tiering function")
```

**Graph 5** visualization of monthly YoY growth for three different price tiers (specified by tier_mils()). 

```python
tier_granular(100)
df_slice["price_tier"] = df_slice.sale_price.map(tier_granular)
fr = pd.DataFrame(list(monthly_yoy_growth(df_slice)))
pretty_plot(fr, "yoy monthly growth using tier_granular tiering function")
```

**Graph 6** visualization of monthly YoY growth for 6 different price tiers as specified by tier_granular().
