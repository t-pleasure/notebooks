## Problem 4

Using the NYC Department of Finance's rolling property sales [data](http://www.nyc.gov/html/dof/html/property/rolling_sales_data.shtml), investigate the following:
* Please produce a timeseries plot showing monthly YoY sales-growth at various price tiers.
  * What has the trend been like for sales of properties priced < \$3M?
  * Modify this to show stats by neighborhood & price-tier; how do sales < \$3M compare for Brooklyn Heights vs. Bedford-Stuyvesant?
---


### General Approach

1. Define tier ranges as {1-3M, 3M+} (todo:)
2. Group rows by (month, year, tier)
3. Compute total number of sales for each (month, year, tier) group.
4. For each month, create a plot consisting of: total sales for each tier plotted against years.

5. Group rows by neighborhood, then repeat 2-4

```python
# import necessary libraries
%matplotlib inline
import csv
import pandas as pd
from pylab import *
from collections import defaultdict

import seaborn as sns
sns.set(context="talk", style="white")
```

```python
class Row:
  def __init__(self, csv_row):
    self.neighborhood = csv_row["NEIGHBORHOOD"]
    self.sale_price = int(csv_row["SALE PRICE"])
    (month, day, year) = csv_row["SALE DATE"].split("/")
    self.sale_month = int(month)
    self.sale_day = int(day)
    self.sale_year = int(year)


def basic_tiers(r):
  return 0 if r.sale_price < 3000000 else 1
    
allrows = (Row(r) for r in csv.DictReader(open("../clean_sales_data.csv")))
rows = [r for r in allrows if r.sale_price > 0]
```

```python
def compute_monthly_yoy(rs, tier_fn, min_year=2003, max_year=2014):
  cnt = defaultdict(int)
  yoy = defaultdict(float)
  seen_tiers = []
  for r in rs:
    tier = tier_fn(r)
    cnt[(r.sale_month, r.sale_year, tier)] += 1
    if tier not in seen_tiers:
      seen_tiers.append(tier)
  for m in range(1,13):
    for yr in range(min_year+1, max_year+1):
      for tr in seen_tiers:
        prev = cnt[(m,yr-1,tr)]
        # TODO IS THIS THE RIGHT WAY TO HANDLE MISSING DATA?
        prev = np.NAN if prev == 0 else prev
        cur = cnt[(m,yr,tr)]
        yoy[(m,yr,tr)] = (cur-prev)/(prev+0.0)
  return yoy

#print sorted(compute_monthly_yoy(rows, basic_tiers).items(), key=lambda ((m,y,t),v):(y,m))
yoy = compute_monthly_yoy(rows, basic_tiers)
#print yoy
def yoy_to_dframe(yoy):
  data = [{"growth": yoy[(mn,yr,tr)], "date": datetime.datetime(yr,mn,1), "tier": tr} \
        for yr in range(2004,2015) for mn in range(1,13) for tr in [0,1] if (yr!=2014 and mn<12)]
  df = pd.DataFrame(data) 
  df.tier.loc[df.tier==0] = "less than 3M tier"
  df.tier.loc[df.tier==1] = "greater than 3M tier"
  return df
df = yoy_to_dframe(yoy)



def plot_dframe(df, graph_title):
  def plot(x, y, data=None, label=None, **kwargs):
    data.plot(x, y, label=label, rot=0, lw=3, grid=False, **kwargs)

  g = sns.FacetGrid(df, hue="tier", size=12,
                  legend_out=False)
  g.map_dataframe(plot, "date", "growth")
  ax = g.axes[0, 0]
  ax.set_title(graph_title, size=14)
  #g.fig.tight_layout()
  ax.legend(loc=(.8, .05))
  return g 

plot_dframe(df, "Monthly year-over-year growth")

#df[df.tier =="less than 3M tier"].plot()
#print cnt
#plt.plot(range(len(data)), data, 'r*')
```

```python

# group-by neighborhood
neigh_groups = defaultdict(list)
for r in rows:
  neigh_groups[r.neighborhood].append(r)

def sort_yoy(yoy):
  return sorted(yoy.items(), key=lambda ((m,y,t),v):(y,m))
    
# compute yoy for brookly
bs_yoy = compute_monthly_yoy(neigh_groups["BEDFORD STUYVESANT"], basic_tiers)
bh_yoy = compute_monthly_yoy(neigh_groups["BROOKLYN HEIGHTS"], basic_tiers)

plot_dframe(yoy_to_dframe(bs_yoy), "bed-stuy")
plot_dframe(yoy_to_dframe(bh_yoy), "brooklyn heights")

#print sort_yoy(bs)
#print sort_yoy(bh)
```
