## Problem 2

Using the NYC Department of Finance's rolling property sales [data](http://www.nyc.gov/html/dof/html/property/rolling_sales_data.shtml), answer the investigate the following:
* What are 2014’s hottest residential neighborhoods, as measured by year-over-year growth in total sale-value and total sale-count?

---


### General Approach

1. Remove rows containing property sale value of $0 (todo: explain).
2. Group rows by neighborhood, then
  2. obtain the total sale-count for each year
  2. obtain the total sale-value for each year
  2. for each year, compute its year-over-year growth by obtaining the difference (in percentage) of the given year compared to its previous year. Do this for sale-count and sale-value.
3. For each neighborhood, determine its year-over-year growth "momentum" by taking the average of its year-over-year growth values.
4. Rank the "hotness" of the neighborhood by the values obtained from (3).

---

### Notes
According to this [glossary](http://www.nyc.gov/html/dof/downloads/pdf/07pdf/glossary_rsf071607.pdf), any property with the class code 1 or 2 is considered to be residential.

```python
# import necessary libraries
%matplotlib inline
import csv
import pandas as pd
from pylab import *
from collections import defaultdict
```

```python
class Row:
    def __init__(self, csv_row):
        self.neighborhood = csv_row["NEIGHBORHOOD"]
        self.tax_class = csv_row["TAX CLASS AT TIME OF SALE"]
        self.sale_price = int(csv_row["SALE PRICE"])  
        (month, day, year) = csv_row["SALE DATE"].split("/")
        self.sale_month = int(month)
        self.sale_day = int(day)
        self.sale_year = int(year)

RES_TAX_CLASS = set(['1','2'])
def is_residential(taxclass):
  return any(map(lambda c: c in RES_TAX_CLASS, taxclass))

def yearly_totals(rows):
  sales_cnt = defaultdict(int)
  sales_value = defaultdict(int)
  for r in rows:
    sales_cnt[r.sale_year] += 1
    sales_value[r.sale_year] += r.sale_price
  return sales_cnt, sales_value

def yoy_growth(yearmap):
  """
  computes year to year growth
  """
  years = sorted(yearmap.keys())
  growth_map = defaultdict(float)
  for yr in years[1:]:
    growth_map[yr] = (yearmap[yr] - yearmap[yr-1])/(yearmap[yr-1] + 1.0)
  return growth_map
```

```python
allrows = (Row(r) for r in csv.DictReader(open("../combined_sales_data.csv")))
rows = [r for r in allrows if r.sale_price > 0 and is_residential(r.tax_class) and r.sale_month!=12]
```

```python
neigh_groups = defaultdict(list)
for r in rows:
  neigh_groups[r.neighborhood].append(r)

cnt_momentum = defaultdict(float)
val_momentum = defaultdict(float)
for (neighborhood, rs) in neigh_groups.items():
  cnt_totals, val_totals = yearly_totals(rs)
  cnt_yoy = yoy_growth(cnt_totals)
  val_yoy = yoy_growth(val_totals)
  cnt_momentum[neighborhood] = np.average(cnt_yoy.values())
  val_momentum[neighborhood] = np.average(val_yoy.values())

cnt_momentum_df = pd.DataFrame(map(lambda (neigh,mu): {"NEIGHBORHOOD": neigh, "SALE_COUNT_MOMENTUM": mu}, cnt_momentum.items()))
val_momentum_df = pd.DataFrame(map(lambda (neigh,mu): {"NEIGHBORHOOD": neigh, "SALE_PRICE_MOMENTUM": mu}, val_momentum.items()))


cnt_momentum_df.sort("SALE_COUNT_MOMENTUM", ascending=False)
```
