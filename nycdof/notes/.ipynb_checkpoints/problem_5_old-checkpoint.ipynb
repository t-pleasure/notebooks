## Problem 5

Using the NYC Department of Finance's rolling property sales [data](http://www.nyc.gov/html/dof/html/property/rolling_sales_data.shtml), investigate the following:
* What neighborhoods have seen the highest turnover in 2014 (as measured by time-between sales of the same property)

---


### General Approach

1. Find all rows that have a 2014 sale date.
2. Group rows (from step 1) by neighborhood
3. For each grouping, find properties (identified by (address, apt, zip)) that have been sold more than once (i.e., indicating turn-over). Keep a count of the number of times the same property has been turned-over for each neighborhood. In addition, record the time(s) between sales of properties.
5. Construct a list of neighborhoods ranked by their number of turn-overs. Use this list to determine neighborhoods with highest turnover.


```python
# import necessary libraries
%matplotlib inline
import csv
import pandas as pd
from pylab import *
from collections import defaultdict
from datetime import datetime,date,timedelta
```

```python
class Row:
    def __init__(self, csv_row):
        self.neighborhood = csv_row["NEIGHBORHOOD"]
        self.sale_price = int(csv_row["SALE PRICE"])
        self.address = csv_row["ADDRESS"]
        self.apt = csv_row["APARTMENT NUMBER"]
        self.zip = int(csv_row["ZIP CODE"])
        (month, day, year) = csv_row["SALE DATE"].split("/")
        self.sale_month = int(month)
        self.sale_day = int(day)
        self.sale_year = int(year)
        self.sale_date = datetime(self.sale_year, self.sale_month, self.sale_day)
        
    def days_since(self, other):
        """
        computes the timedelta.days of this row's sale date and other's sale date.
        Returns a negative number if other's sale date is after self's sale date.
        """
        return (self.sale_date - other.sale_date).days
    
allrows = (Row(r) for r in csv.DictReader(open("../combined_sales_data.csv")))
rows = [r for r in allrows if r.sale_price > 0 and r.sale_year == 2014]
```

```python
def turnover_agg(rows):
  # group rows by address tuple (addr, apt, zip)
  addr_groups = defaultdict(list)
  for r in rows:
    addr_groups[(r.address, r.apt, r.zip)].append(r)
  # remove any groupings with only one elements (indicative of no turn over at addr)
  return dict((k,rs) for (k, rs) in addr_groups.items() if len(rs)>1)

def rows_to_turnover_times(rows):
  """
  given a list of rows, compute the turnover times between rows.
  """
  srows = sorted(rows, key=lambda r: r.sale_date)
  tov_times = []
  for ind in xrange(1, len(srows)):
    days_delta = srows[ind].days_since(srows[ind-1])
    tov_times.append(days_delta)
  return tov_times
    
def turnover_stats(turnover_groups):
  # count number of turn-overs
  n_tov = 0
  total_tov = 0.0
  for (addr, rs) in turnover_groups.items():
    n_tov += (len(rs) - 1)
    tov_times = rows_to_turnover_times(rs)
    total_tov += sum(tov_times)
    
  return n_tov, total_tov/n_tov, 1


neigh_groups = defaultdict(list)
for r in rows:
  neigh_groups[r.neighborhood].append(r)

#print neigh_groups.keys()

stats = []
for (neigh, rs) in neigh_groups.items():
  tgroups = turnover_agg(rs)
  if tgroups:
    (tot_to, avg_to, _) = turnover_stats(tgroups)
    if avg_to > 0:
      stats.append((neigh, tot_to, avg_to))

print "\n**by average turnover time / total turnovers"
for (n,tot,mu) in sorted(stats, key=lambda (n,tot,mu): mu/tot)[:10]:
  print n, tot, mu
    
print "\n**by lowest average turnover time"
for (n,tot,mu) in sorted(stats, key=lambda (n,tot,mu): mu)[:10]:
  print n, tot, mu
    
print "\n**by highest number of turnovers"
for (n,tot,mu) in sorted(stats, key=lambda (n,tot,mu): tot, reverse=True)[:10]:
  print n, tot, mu
    
```
