Using the NYC Department of Finance's rolling property sales [data](http://www.nyc.gov/html/dof/html/property/rolling_sales_data.shtml), investigate the following:
* Does the residential market prefer new-construction (buildings built within 2years of the sale date)? How does 2014 compare with previous years in this regard?

---


### General Approach

#### Main Idea:
* We will determine whether or not the residential market place prefers new-constructions by evaluating whether or not they would on average **pay more** for a new home than one that has the same *building-class* and *zip code* but is not considered a "new construction."


#### Steps:
1. Remove transfer of ownership rows (e.g., sale price == $0)
2. Only look at Residential properties.
3. Group the rows into "built within 2 years of sale date" and "NOT built within 2 years of sale date"
4. For each group, group by (residential type, zip code) and compute average sale price
5. Repeat 1-4 for for individual years.


## Data Analysis

```python
# import necessary libraries
%matplotlib inline
import csv
import pandas as pd
import datetime as dt
from pylab import *
```

### Reading and cleaning data

#### We will need to:
1. **remove** rows with $0 sale price (these are "transfer of ownership" events)
2. **keep** rows that represent residential properties (i.e., rows with tax class 1 or 2)
3. **remove** rows that do not have valid **year built** fields (i.e., rows with year_built=0)

```python
# read csv data to pandas DataFrame
df = pd.read_csv("../clean_sales_data.csv")

# add additional fields and rename fields to make life easier
df["sale_price"] = df["SALE PRICE"]
df["zip"] = df["ZIP CODE"]
df["tax_class"] = df["TAX CLASS AT TIME OF SALE"]
df["building_class"] = df["BUILDING CLASS AT TIME OF SALE"]
df["sale_date"] = df["SALE DATE"].map(lambda s: dt.datetime.strptime(s, "%m/%d/%Y"))
df["sale_day"] = df.sale_date.map(lambda d: d.day)
df["sale_month"] = df.sale_date.map(lambda d: d.month)
df["sale_year"] = df.sale_date.map(lambda d: d.year)
df["year_built"] = df["YEAR BUILT"]
df["is_new_constr"] = df["year_built"].map(lambda y: y + 2) >= df["sale_year"]

# remove columns that are not needed
df_slice = df[["sale_price","zip","tax_class","building_class","sale_date",
               "sale_day","sale_month","sale_year","year_built","is_new_constr"]]

## PERFORM FILTERING ##
# 1. remove transfer of ownership rows
df_slice = df_slice[df_slice.sale_price > 0]
# 2. keep only residential properties rows
df_slice = df_slice[(df_slice.tax_class==1) | (df_slice.tax_class==2)]
# 3. keep rows with year_built information (year_built>0)
df_slice = df_slice[df_slice.year_built >0]
```

## Aggregating "similar homes"
This section will compute the statistics necessary for us to determine home purchasing preferences. More specifically, this section will compute average sale prices for "new" and "old" homes constructions for particular (zip_code, building_class) tuples.

```python
def construct_comparison(fr):
  """
  Given a DataFrame with cols ["building_class","zip","sale_price","is_new_constr"],
  this function will use the input frame to construct a new data frame with:
   ** building class
   ** zip code
   ** average selling prices for "OLD home sales" with (zip code, building class)
   ** average selling prices for "NEW home sales" with (zip code, building class)
  Input:
    * fr (pandas.DataFrame) -- input dataframe to compute stats over
  Output:
    * (pandas.DataFrame) -- DataFrame containing stats for new and old home sales.
  """
  # GROUP DATA INTO (building class, zip, is_new_constr)
  # then compute average sale prices of each group
  grp = fr.groupby(["building_class","zip","is_new_constr"])["sale_price"].mean()

  old_prices = {} 
  new_prices = {}
  for ((bclass, zipcode, isnew), mu) in grp.iteritems():
    d = new_prices if isnew else old_prices
    d[(bclass,zipcode)] = mu

  # find (building-class,zipcode) that have both stats for old and new properties
  valid_keys = set(new_prices.keys()).intersection(old_prices.keys())
  tmp = [{"building_class": bclass,
         "zip": zipcode,
         "avg_price_old": old_prices[(bclass, zipcode)],
         "avg_price_new": new_prices[(bclass, zipcode)]} for (bclass, zipcode) in valid_keys]
  
  return pd.DataFrame(tmp)
```

#### A peek into old vs new sale prices
The following will give us a peak into price differences bettween old and new home construction sales.

```python
stats_fr = construct_comparison(df_slice)
stats_fr[["building_class","zip","avg_price_old","avg_price_new"]].head(15)
```

**Table2** A sample of the data frame containing price differences between old and new constructions based on (zipcode, building class) tuples.

This section will figure what types of buildings+zip have higher sale prices than their newer counterpart.

### Are there more new home sales with higher prices than older home sales?

```python
def old_vs_new(sfr):
  n_old = len(sfr[sfr.avg_price_old > sfr.avg_price_new])
  n_new = len(sfr[sfr.avg_price_old < sfr.avg_price_new])
  return (n_old, n_new)

cnt_old, cnt_new = old_vs_new(stats_fr)
pd.DataFrame([{"old_pref":cnt_old, "new_pref":cnt_new, "year":"2003-2014"}])
```

**Table 5** Counts of types that prefer old constructions and which prefer new constructions as evaluated over data from 2003-2014. 


### What does the preference look like for individuals years?

```python

agg = []
for (yr,gfr) in df_slice.groupby("sale_year"):
  o,n = old_vs_new(construct_comparison(gfr))
  agg.append({"old_pref":o, "new_pref":n, "year":yr})
  
pd.DataFrame(agg)
```

**Table 6** Counts of types that prefer old constructions and which prefer new constructions as evaluated over individual years.


## Conclusion
* Is there a preference for new home constructions? -- **Yes!**
* How does 2014 compare to other years? -- The market place in 2014 has a preference for new constructions. In addition, **every year prior to 2014 has also shown that the market prefers new home constructions.** In fact, when comparing the number of zipcode+building_class types that yield higher sale prices for new construction homes, there are roughly ~3x the number of types that prefer new constructions over old ones.
